```python

import kagglehub
import pandas as pd
import os

# Datasetni yuklab olamiz
path = kagglehub.dataset_download("naveenkumar20bps1137/sample-superstore")
print("Fayllar joylashgan manzil:", path)
```

```python
# --- Task 1: Loading the Data ---
# Identify the filename in the downloaded directory
files = os.listdir(path)
print(f"Files found in directory: {files}")

# Define the full path (Assuming the file is 'SampleSuperstore.csv')
# Note: encoding='windows-1252' is often needed for this specific dataset
file_path = os.path.join(path, files[0])
df = pd.read_csv(file_path, encoding='latin-1')

# Display the first few rows
print("Data successfully loaded:")
display(df.head())

# --- Task 2: Inspecting the Structure ---
print("\n--- Dataset Info ---")
df.info()

print("\n--- Missing Values Check ---")
print(df.isnull().sum())

print("\n--- Descriptive Statistics ---")
display(df.describe())
```

```python
# Har bir ustunda nechta bo'sh joy borligini tekshiramiz
missing_counts = df.isnull().sum()
print("Bo'sh joylar soni:\n", missing_counts)

# Agar bo'sh joylar bo'lsa, ularni to'ldirish (Professional usul):
# Raqamli ustunlarni median (o'rtacha) bilan to'ldiramiz
for col in df.select_dtypes(include=['number']).columns:
    df[col] = df[col].fillna(df[col].median())

# Matnli ustunlarni eng ko'p takrorlangan qiymat (mode) bilan to'ldiramiz
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nTozalashdan keyin bo'sh joylar qolmadi.")
```

```python
# Dublikatlarni sanaymiz
duplicates = df.duplicated().sum()
print(f"Topilgan dublikatlar soni: {duplicates}")

# Dublikatlarni o'chiramiz
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Dublikatlar o'chirildi. Yangi jadval o'lchami: {df.shape}")
```

```python
# 1. Har bir mijoz uchun jami savdoni hisoblaymiz
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

# 2. Har bir mijoz necha marta buyurtma berganini hisoblaymiz
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

# 3. Har bir mijozning o'rtacha savdosini hisoblaymiz
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# 4. Hamma ma'lumotlarni bitta jadvalga birlashtiramiz (Merge)
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')

# Natijani ko'ramiz
print("Customer Summary Table:")
display(customer_summary.head(10))

# 5. TOZALANGAN ma'lumotlarni CSV fayl qilib saqlaymiz
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

print("\nTabriklayman! Tozalangan fayllar 'superstore_cleaned.csv' va 'customer_summary.csv' nomi bilan saqlandi.")
```

## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Fill missing numerical values with median (robust to outliers)
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

## Task 3: Remove Duplicates (pre-built)

```python
# Check for duplicates
n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```
